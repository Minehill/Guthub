```python
import cv2 as cv
import numpy as np
def nothing():
    pass

#设置滑块窗口
cv.namedWindow('BGR')
cv.resizeWindow('BGR',600,300)
cv.createTrackbar('B_L','BGR',0,255,nothing)
cv.createTrackbar('B_H','BGR',0,255,nothing)
cv.createTrackbar('G_L','BGR',0,255,nothing)
cv.createTrackbar('G_H','BGR',0,255,nothing)
cv.createTrackbar('R_L','BGR',0,255,nothing)
cv.createTrackbar('R_H','BGR',0,255,nothing)
while(1):
    B_H = cv.getTrackbarPos('B_H','BGR')
    B_L = cv.getTrackbarPos('B_L','BGR')
    G_H = cv.getTrackbarPos('G_H','BGR')
    G_L = cv.getTrackbarPos('G_L','BGR')
    R_H = cv.getTrackbarPos('R_H','BGR')
    R_L = cv.getTrackbarPos('R_L','BGR')

#读取目标图像
    img = cv.imread('blue.png')
    img = cv.resize(img,(-1,-1),fx=0.5,fy=0.5)

#定义阈值
    lower_ = np.array([B_L,G_L,R_L])
    upper_ = np.array([B_H,G_H,R_H])
#设置掩膜并与原图相加
    mask = cv.inRange(img,lower_,upper_)
    fina = cv.bitwise_and(img,img,mask=mask)

    cv.imshow('img',img)
    cv.imshow('mask',mask)
    cv.imshow('fina',fina)
    k = cv.waitKey(5) & 0xFF
    if k == 27:
        break
cv.destroyAllWindows()

```
