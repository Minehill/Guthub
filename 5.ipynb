```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt


im = cv.imread('./blue.png')
#im = cv.GaussianBlur(im,(3,3),0)
plt.imshow(im)
plt.show()
img = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
#img = cv.resize(img, (-1, -1), fx=0.5, fy=0.5)
ret, thresh = cv.threshold(img,200,255,cv.THRESH_BINARY) #该函数返回一共两个内容，第一个为一个浮点数，第二个为一个表示图像的数组
#plt.figure(figsize=(12,10))
plt.imshow(thresh,cmap='gray')
plt.show()
kernel = np.ones((4, 4), np.uint8)
dilation = cv.dilate(thresh, kernel, iterations=1)
#plt.figure(figsize=(12,10))
plt.imshow(dilation,cmap='gray')
plt.show()
ret, dilation = cv.threshold(dilation,250,255,cv.THRESH_BINARY)
kernel_ = np.ones((4, 4), np.uint8)
erosion=cv.erode(dilation,kernel_,iterations=1)
#plt.figure(figsize=(12,10))
plt.imshow(erosion,cmap='gray')
plt.show()
contours, hierarchy = cv.findContours(erosion, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
cv.drawContours(im,contours,-1,(0,0,255),3)
#plt.figure(figsize=(12,10))
#plt.imshow(cv.cvtColor(im,cv.COLOR_BGR2RGB))
plt.imshow(im)
plt.show()
area1 = 10
area2 = 30
newcontours = []
for cnt in contours:
    (x,y),radius = cv.minEnclosingCircle(cnt)
    center = (int(x),int(y))
    radius = int(radius)
    cv.circle(im,center,radius,(0,255,0),2)
    if radius >= area1 and radius <= area2:
        newcontours.append(cnt)
cv.drawContours(im, newcontours, -1, (0, 255, 0), 3)
cv.imshow('im',im)
cv.waitKey(0)
plt.imshow(im)
plt.show()



    
#for cnt in contours:
 #   x,y,w,h = cv.boundingRect(cnt)
  #  font = cv.FONT_HERSHEY_COMPLEX
    #cv.putText(im,f'{int(w/h)}',{x,y},font,1,(0,0,255),1,)
   # cv.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

#cv.imshow('img',img)
#cv.waitKey(0)
```
